```python
import pandas as pd
```

```python
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
data=pd.read_csv("C:/Users/Abhijit Kharade/Downloads/Association rule/book.csv")
```

```python
data.head()
```

```python
data.info()
```

```python
data.shape
```

```python
frequent_dataset=apriori(data,min_support=0.1,use_colnames=True)
```

```python
frequent_dataset
```

```python
rules=association_rules(frequent_dataset,metric='lift',min_threshold=0.7)
```

```python
rules
```

```python
rules.sort_values('lift',ascending=False)[0:20]
```

```python
rules[rules.lift>1]
```
